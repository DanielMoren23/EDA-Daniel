```python
df_filtered["Year"] = df_filtered["GAME_DATE"].dt.year
df_avg_points = df_filtered.groupby("Year").apply(
    lambda x: (x["PTS_home"] + x["PTS_away"]).mean()
).reset_index(name="Avg_Points_Per_Game")
print(df_avg_points)
```

```python
sns.set_style("whitegrid", {"grid.color": "#e0e0e0"})  
sns.set_palette("viridis") 
plt.figure(figsize=(12, 6))
ax = sns.lineplot(
    data=df_avg_points,
    x="Year",
    y="Avg_Points_Per_Game",
    marker="o",
    linewidth=2.5,
    markersize=8,
    color="#2e86c1"  
)
plt.title("Evolución de los Puntos Promedio por Partido (NBA: 2004-2025)", fontsize=16, pad=20)
plt.xlabel("Año", fontsize=12, labelpad=10)
plt.ylabel("Puntos Promedio", fontsize=12, labelpad=10)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle="--", alpha=0.7)
max_year = df_avg_points.loc[df_avg_points["Avg_Points_Per_Game"].idxmax()]
ax.annotate(f'Máximo: {max_year["Avg_Points_Per_Game"]:.1f} pts',
            xy=(max_year["Year"], max_year["Avg_Points_Per_Game"]),
            xytext=(max_year["Year"]-2, max_year["Avg_Points_Per_Game"]+5),
            arrowprops=dict(arrowstyle="->", color="black"))
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df_avg_points["Avg_Points_Per_Game"], bins=10, kde=True, color="blue")
plt.title("Distribución de Puntos Promedio por Año")
plt.xlabel("Puntos Promedio")
plt.ylabel("Frecuencia")
```

```python
df_home = df[['date', 'home_team', 'away_team', 'home_score', 'away_score']].copy()
df_home['date'] = pd.to_datetime(df_home['date'])
df_home['year'] = df_home['date'].dt.year
df_home['home_result'] = df_home.apply(lambda row: 'win' if row['home_score'] > row['away_score'] 
                                       else 'loss' if row['home_score'] < row['away_score'] 
                                       else 'draw', axis=1)
df_home['decade'] = (df_home['year'] // 10) * 10
home_win_rate = df_home.groupby('decade')['home_result'].apply(lambda x: (x == 'win').mean())
print(home_win_rate)
```

```python
df_home['is_draw'] = df_home['home_result'] == 'draw'
draw_rate = df_home.groupby('decade')['is_draw'].mean()
print(draw_rate)
```

```python
df = pd.read_csv("Partidos desde 1872 hasta 2017.csv")
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df_filtered = df[['year', 'home_score', 'away_score']].copy()
df_filtered['home_result'] = df_filtered.apply(
    lambda row: 'win' if row['home_score'] > row['away_score'] 
    else 'loss' if row['home_score'] < row['away_score'] 
    else 'draw', axis=1
)
df_filtered['decade'] = (df_filtered['year'] // 10) * 10
home_win_rate = df_filtered.groupby('decade')['home_result'].apply(lambda x: (x == 'win').mean())
draw_rate = df_filtered.groupby('decade')['home_result'].apply(lambda x: (x == 'draw').mean())

# Gráfico 1: Evolución de la ventaja de jugar en casa 
plt.figure(figsize=(10, 5))
plt.plot(home_win_rate.index, home_win_rate.values, marker='o', linestyle='-', color='b', label='Victorias en casa')
plt.xlabel('Década')
plt.ylabel('Proporción de victorias en casa')
plt.title('Evolución de la ventaja de jugar en casa')
plt.grid(True)
plt.legend()

# Gráfico 2: Evolución de la frecuencia de empates
plt.figure(figsize=(10, 5))
plt.plot(draw_rate.index, draw_rate.values, marker='s', linestyle='-', color='r', label='Empates')
plt.xlabel('Década')
plt.ylabel('Proporción de empates')
plt.title('Evolución de la frecuencia de empates en partidos')
plt.grid(True)
plt.legend()
```

```python
try:
    data = pd.read_csv("boxing stats.csv") 
    print("Dataset cargado correctamente.")
except FileNotFoundError:
    print("Error: El archivo no se encuentra en la ruta especificada.")
    exit()

print("Columnas disponibles:", data.columns.tolist())
data = data[['f_boxer', 'f_boxer_won', 'f_boxer_lost', 'f_boxer_drawn', 'f_boxer_KOs']]

# Calcular el total de peleas y el porcentaje de victorias
data['total_fights'] = data['f_boxer_won'] + data['f_boxer_lost'] + data['f_boxer_drawn']
data['win_rate'] = (data['f_boxer_won'] / data['total_fights']) * 100
print(data.head())
```

```python
bins = [0, 10, 20, 30, 40, 50, 100, 200]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-100', '101-200']
data['fight_interval'] = pd.cut(data['total_fights'], bins=bins, labels=labels)

# Calcular promedio de porcentaje de victorias por intervalo
win_rate_by_interval = data.groupby('fight_interval')['win_rate'].mean().reset_index()
print(win_rate_by_interval)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(win_rate_by_interval['fight_interval'], win_rate_by_interval['win_rate'], color='skyblue', edgecolor='black')
plt.xlabel("Intervalo de peleas")
plt.ylabel("Porcentaje promedio de victorias (%)")
plt.title("Relación entre número de peleas y porcentaje de victorias")
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
try:
    data = pd.read_csv("boxing stats.csv")  
    print("Dataset cargado correctamente.")
except FileNotFoundError:
    print("Error: El archivo no se encuentra en la ruta especificada.")
    exit()

print("Columnas disponibles:", data.columns.tolist())
columnas_deseadas = ['f_boxer_won', 'f_boxer_lost', 'f_boxer_drawn']
data = data.dropna(subset=columnas_deseadas)  
data[columnas_deseadas] = data[columnas_deseadas].astype(int)  
data = data[data[columnas_deseadas].apply(lambda row: row.sum() > 0, axis=1)]  
if 'f_boxer' in data.columns:
    data['country'] = data['f_boxer'].apply(lambda x: x.split()[-1])
else:
    print("Error: No hay una columna que represente el nombre del boxeador. Imposible calcular el país.")
    exit()
# Calcular el total de peleas y el porcentaje de victorias
data['total_fights'] = data['f_boxer_won'] + data['f_boxer_lost'] + data['f_boxer_drawn']
data['win_rate'] = (data['f_boxer_won'] / data['total_fights']) * 100
print(data.head())
```

```python
country_win_rate = data.groupby('country')['win_rate'].mean().sort_values(ascending=False).reset_index()
print(country_win_rate)
```

```python
plt.figure(figsize=(12, 6))
top_countries = country_win_rate.head(10)
plt.bar(top_countries['country'], top_countries['win_rate'], color='skyblue', edgecolor='black')
plt.xlabel("Países")
plt.ylabel("Porcentaje promedio de victorias (%)")
plt.title("Tasa de éxito promedio por país (Top 10)")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
```

```python
try:
    data = pd.read_csv("boxing stats.csv")  
    print("Dataset cargado correctamente.")
except FileNotFoundError:
    print("Error: El archivo no se encuentra en la ruta especificada.")
    exit()
print("Columnas disponibles:", data.columns.tolist())
columnas_deseadas = ['f_boxer_age', 'f_boxer_won', 'f_boxer_lost', 'f_boxer_drawn']
data = data.dropna(subset=columnas_deseadas)  
data[columnas_deseadas] = data[columnas_deseadas].astype(int)  
data = data[data[columnas_deseadas].apply(lambda row: row.sum() > 0, axis=1)]  
data = data[(data['f_boxer_age'] >= 18) & (data['f_boxer_age'] <= 50)]

# Calcular el total de peleas y el porcentaje de victorias
data['total_fights'] = data['f_boxer_won'] + data['f_boxer_lost'] + data['f_boxer_drawn']
data['win_rate'] = (data['f_boxer_won'] / data['total_fights']) * 100
print(data.head())
```

```python
bins = [18, 25, 30, 35, 40, 45, 50]
labels = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-50']
data['age_group'] = pd.cut(data['f_boxer_age'], bins=bins, labels=labels, right=False)

# Agrupar por rango de edad y calcular la tasa de éxito promedio
age_win_rate = data.groupby('age_group')['win_rate'].mean().reset_index()
print(age_win_rate)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(age_win_rate['age_group'], age_win_rate['win_rate'], color='skyblue', edgecolor='black')
plt.xlabel("Rangos de edad")
plt.ylabel("Porcentaje promedio de victorias (%)")
plt.title("Tasa de éxito promedio por rango de edad")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
```
